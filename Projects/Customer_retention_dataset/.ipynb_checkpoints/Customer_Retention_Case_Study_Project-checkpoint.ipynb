## From the above verification using value_counts() method and also using histplot we can see clear that the data is unbalanced in the both classes where the custmer agreed[1] is majority class and custmer agreed[0] is minority class...

```python
## let's solve the unbalance problem

## importang the libraries
from imblearn.over_sampling import SMOTE

SM = SMOTE()
```

```python
## let's first split our dataset into label and features.

y = data['47 Getting value for money spent']

x = data.iloc[:,:44]
```

```python
x_over , y_over = SM.fit_resample(x, y)
```

```python
y_over.hist(color = 'y',alpha = 0.7, figsize=(10,10))
```

### __Now we can see that our both classes of the label are balanced...__
