### Q1. Scrape the details of most viewed videos on YouTube from Wikipedia:
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ You need to find following details:
### A) Rank
### B) Name
### C) Artist
### D) Upload date E) Views

```python
from tqdm import tqdm
## this will help track the for loops.
```

```python
driver = webdriver.Chrome('/Users/olivierla412/Desktop/chromedriver')
time.sleep(7)
## fetching the wibsite
wiki_url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url=wiki_url)
time.sleep(3)
videos_Xpath1 = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td')
time.sleep(2)
table1_details = []
for i in videos_Xpath1:
    table1_details.append(i.text)
time.sleep(2)

## creating some empty list.

Rank =[]
Video_Name =[]
Artist_Name = []
Upload_date =[]
Views =[]

for x in tqdm(range(0,len(table1_details),6)):
    Rank.append(table1_details[x])
    Video_Name.append(table1_details[x+1])
    Artist_Name.append(table1_details[x+2])
    Views.append(table1_details[x+3])
    Upload_date.append(table1_details[x+4])
### dealing now with the second table on the website

time.sleep(3)
videos_Xpath2 = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[5]/tbody/tr/td')
table2_details =[]
for i in videos_Xpath2:
    table2_details.append(i.text)
time.sleep(2)
for x in tqdm(range(0,len(table2_details),5)):
    Video_Name.append(table2_details[x])
    Artist_Name.append(table2_details[x+1])
    Views.append(table2_details[x+2])
    Upload_date.append(table2_details[x+3])
    Rank.append('na') ## the second table videos don't have rank
time.sleep(2)  


### dealing now with the Third table on the website

table3_details =[]
videos_Xpath3 = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[6]/tbody/tr/td')
for x in videos_Xpath3:
    table3_details.append(x.text)    
for x in tqdm(range(0,len(table3_details),8)):
    Video_Name.append(table3_details[x])
    Artist_Name.append(table3_details[x+1])
    Views.append(table3_details[x+2])
    Upload_date.append(table3_details[x+3])
    Rank.append('na') ## the second table videos don't have rank
    
data1 = pd.DataFrame({
    'Rank':Rank,
    'Video_Name':Video_Name,
    'Artist_Name':Artist_Name,
    'Upload_date':Upload_date,
    'Views':Views 
})    
pd.options.display.max_rows = 8    
data1
```

```python
## setting the driver
driver = webdriver.Chrome('/Users/olivierla412/Desktop/chromedriver')
time.sleep(3)
## fetching the website
url_bill = ' https://www.billboard.com/'
driver.get(url=url_bill)
time.sleep(5)
driver.maximize_window()
time.sleep(2)
driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/nav/ul/li[3]/a').click()
for _ in range(500):
    driver.execute_script("window.scrollBy(0,10000)") ### scrolling the window till the bottom of the page.

time.sleep(3)
name = driver.find_elements_by_xpath('//*[@id="charts"]/div/div/div/ol/li/button/span/span')
#Make empty lists   
music_dict = {}
music_dict['Song_name']= []
music_dict['Artist_name']= []
music_dict['Last week rank']= []
music_dict['Peak rank']= []
music_dict['Weeks on board']= []

My_List = []
for x in name:
    My_List.append(x.text)
for i in range(0,len(My_List),5):
    music_dict['Song_name'].append(My_List[i+2])
    music_dict['Artist_name'].append(My_List[i+3])
    
time.sleep(2)    
MyList2 = []
ranks = driver.find_elements_by_xpath('//*[@id="charts"]/div/div[7]/div/ol/li/button/div')
for i in ranks:
    MyList2.append(i.text.replace('\n',' ').split(sep=' ',maxsplit=3))
for x in MyList2:
    music_dict['Last week rank'].append(x[0])
    music_dict['Peak rank'].append(x[1])
    music_dict['Weeks on board'].append(x[2])
## creating the data frame.

Music_df = pd.DataFrame.from_dict(music_dict)
Music_df 
```

```python
### fetching the website
UCI_url ='https://archive.ics.uci.edu/'
driver.get(UCI_url)
time.sleep(2)
driver.maximize_window()
time.sleep(5)
driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b').click()
### creating some empty list
time.sleep(3)
Datasets_dict = {}
Datasets_dict['Dataset name'] =[]
Datasets_dict['Data type'] =[]
Datasets_dict['Task'] =[]
Datasets_dict['Attribute type'] =[]
Datasets_dict['No of instances'] =[]
Datasets_dict['No of attribute'] =[]
Datasets_dict['Year'] =[]
table_Xpath = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td')
time.sleep(2)
## creating an empty list
data_list = []
for x in table_Xpath:
    data_list.append(x.text)

for x in tqdm(range(7,len(data_list),7)):
    Datasets_dict['Dataset name'].append(data_list[x])
    Datasets_dict['Data type'].append(data_list[x+1])
    Datasets_dict['Task'].append(data_list[x+2])
    Datasets_dict['Attribute type'].append(data_list[x+3])
    Datasets_dict['No of instances'].append(data_list[x+4])
    Datasets_dict['No of attribute'].append(data_list[x+5])
    Datasets_dict['Year'].append(data_list[x+6])
## creating the data frame
Details_of_Datasets_df = pd.DataFrame.from_dict(Datasets_dict)
Details_of_Datasets_df
```
